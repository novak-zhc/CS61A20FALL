## 🖋️ 一、环境图绘制规则与词法作用域

### 1. 函数定义时 (Definition Time)

> [!INFO] **函数定义时的处理**
> * **创建函数值：** 生成一个包含函数代码、形参列表、和**父框架**引用的函数对象。
> * **父框架引用：** 函数对象**记住**了它被定义时的环境。这是**词法作用域 (Lexical Scoping)**的基础。
> * **名称绑定：** 将函数名绑定到当前框架中的这个函数对象。
> * **特点：** 此时不执行函数体内的代码，只进行初步语法检查（惰性求值）。

### 2. 函数执行时 (Execution Time)

> [!IMPORTANT] **函数执行时的步骤**
> * **创建局部框架：** 生成一个新的局部框架，标题为函数名。
> * **设置父框架（关键）：** 新框架的父框架被设置为函数对象中记录的**父框架**（定义时的环境），**而非调用它的环境**。
> * **参数绑定：** 将实参值绑定到局部框架中的形参名。
> * **执行函数体：** 开始执行函数体内代码，**此时才进行变量查找和函数调用检查**。

### 3. 闭包与查找链

> [!EXAMPLE] **闭包示例：`make_adder(5)(3)`**
> ```python
> def make_adder(n):
>     def adder(k):
>         return n + k  # 查找 n
>     return adder
> ```
> * **`adder` 的父框架**：是 `make_adder(5)` 的局部框架（`f1`）。
> * **查找 `n`**：当 `adder(3)` 执行 `n + k` 时，它会沿着父框架链，回到 `f1` 框架中找到 `n=5`。

## 💻 二、代码求值与执行顺序

### 1. 函数定义与执行的时序

> [!WARNING] **Python 的惰性求值**
> * 在 `f` 的定义中引用未定义的 `g` 是**允许的**，因为定义时不检查。
> * 只有在 **执行** `f()` 时，Python 才会查找 `g`。
> * **错误点：** 函数 `g` 如果引用了其作用域链中不存在的变量，则会在执行时抛出 `NameError`。

### 2. 并行赋值的执行顺序

> [!INFO] **`pred, curr = curr, curr + pred`**
> 1.  **求值右侧 (RHS)——原子性：** 解释器计算右侧所有表达式 (`curr` 和 `curr + pred`) 的值，并存储在一个**临时元组**中。
> 2.  **同时赋值 (LHS)：** 解释器使用临时元组中的值，同时赋值给左侧的 `pred` 和 `curr`。
> * **目的：** 确保右侧的表达式使用的是赋值操作**之前**的旧值。

### 3. 柯里化与高阶函数

> [!IMPORTANT] **`make_adder(1)(2)` 模式**
> * **概念：** 这种模式体现了**柯里化 (Currying)** 的核心思想，即把多参数函数转化为单参数函数的链式调用。
> * **执行：**
>     1.  `make_adder(1)` 执行：返回一个**新的函数对象** (`adder`)。
>     2.  `(2)` 执行：使用上一步返回的函数对象作为操作符，传入 `2` 作为参数，进行第二次函数调用。
